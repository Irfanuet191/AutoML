```python
#CNN network development code 
import numpy as np
import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow import keras
import cv2
import glob
import PIL
import PIL.Image
import matplotlib.pyplot as plt

#DATASET USED IS CIFAR10
(dataTrain,dataTest),info=tfds.load('cifar10', split=['train','test'], with_info=True,as_supervised=True)

#Named automl for cnn network development class
class automl:
  
  def __init__(self):
    self.model1 =  keras.Sequential()
    self.test_data=None
    self.train_data=None
    self.model_history=None
   
  def add_layer_Input(self,shape_img=(32,32,3)):
     self.model1.add(keras.Input(shape=shape_img))
    
  def add_layer_MaxPooling2D(self,pool_size_kernal=(2,2)):
    self.model1.add(keras.layers.MaxPooling2D(pool_size=pool_size_kernal))

  def add_layer_Conv2D(self,filters=64,kernal_size=(3,3),activation_func='relu',padding_status='valid'):
     self.model1.add(keras.layers.Conv2D(filters,kernal_size,padding=padding_status,activation=activation_func))
  
  def add_layer_Flatten(self):
    self.model1.add(keras.layers.Flatten())
  
  def add_layer_Dense(self,neuronsNum=64,activation_func='relu'):
    self.model1.add(keras.layers.Dense(neuronsNum,activation=activation_func))
  
  def summary_model(self):
    self.model1.summary()

  def model_train_test(self,Batch_Size=64,Epoch_No=10):
    self.model1.compile(
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=keras.optimizers.Adam(learning_rate=3e-4),
    metrics=["accuracy"])
    self.history_Model=self.model1.fit(self.test_data,batch_size=64,epochs=10,verbose=2,validation_data=self.train_data)
    return self.history_Model.history['accuracy'],self.history_Model.history['val_accuracy']

  def plot_results(self):
    plt.plot(self.model_history.history['accuracy'])
    plt.plot(self.model_history.history['val_accuracy'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'val'], loc='upper left')
    plt.show()

    plt.plot(self.model_history.history['loss'])
    plt.plot(self.model_history.history['val_loss'])
    plt.title('model lossess')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'val'], loc='upper left')
    plt.show()
 

  def Normalize(self):
      def normalize_img(image, label):
        return tf.cast(image, tf.float32) / 255., label

      self.train_data = dataTrain.map(
          normalize_img, num_parallel_calls=tf.data.AUTOTUNE)
      self.train_data= self.train_data.cache()
      # dataTrain_1 = dataTrain_1.shuffle(info.splits['train'].num_examples)
      self.train_data = self.train_data.batch(64)
      self.train_data = self.train_data.prefetch(tf.data.AUTOTUNE)

      self.test_data = dataTest.map(
          normalize_img, num_parallel_calls=tf.data.AUTOTUNE)
      self.test_data = self.test_data.batch(64)
      self.test_data = self.test_data.prefetch(tf.data.AUTOTUNE)
```

```python
#ANN Network development Class
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

dataset = pd.read_csv('/content/drive/MyDrive/Churn_Modelling.csv')
X = dataset.iloc[:, 3:-1].values
y = dataset.iloc[:, -1].values

#Label Encoding Gender Colmun
le = LabelEncoder()
X[:, 2] = le.fit_transform(X[:, 2])

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(ct.fit_transform(X))

class ANN:
  
  def __init__(self):
    
    self.model2 =  keras.Sequential()
    self.X_train=None
    self.X_test=None
    self.y_train=None
    self.y_test=None

  def get_test_train_data_split(self,split_ratio=0.2):
    self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(X, y, test_size = split_ratio, random_state = 0)
    self.X_train = sc.fit_transform(self.X_train)
    self.X_test = sc.transform(self.X_test)  

  #Method to make input layer
  def add_Input_and_FHL(self,neurons=12,activation_func='relu'):
     self.model2.add(keras.layers.Dense(units=neurons, activation=activation_func, input_shape=self.X_train[0].shape))
   
  def add_layer_Dense(self,neuronsNum=64,activation_func='relu'):
    self.model2.add(keras.layers.Dense(neuronsNum,activation=activation_func))
 
  def summary_model(self):
    self.model2.summary()

  def model_train_test(self,Batch_Size=32,Epoch_No=100):
    self.model2.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    #global history_Model
    self.history_Model=self.model2.fit(self.X_train,self.y_train, batch_size = Batch_Size, epochs = Epoch_No,validation_data=(self.X_test,self.y_test))
    # self.history_Model=self.model1.fit(self.test_data,batch_size=64,epochs=10,verbose=2,validation_data=self.train_data)
  def model_Predict(self,Array,Threshold):    
    return (self.model2.predict(sc.transform([Array]))> Threshold)
  def model_trained_history_info(self):
     return self.history_Model.history['accuracy'],self.history_Model.history['val_accuracy'],self.history_Model.history['loss'],self.history_Model.history['val_loss']
```
